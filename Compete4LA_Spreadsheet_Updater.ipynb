Purpose of this program is to classify all of the businesses we have in our small business database by Hub Zone and Valley Regions.
During our previous data cleaning, we lost our latitude, longitude, census tract, and GeoMatch ID. The purpose of this code is to 
recover that.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

 # Download respective spreadsheets anch check if they load correctly.

```python
spreadsheet_name = "C:/Users/User/Downloads/Excel Spreadsheets/USC Center for Economic Development/Compete4LA_Compiled_Spreadsheet_With_Surveys 3.21.23.csv"
new_spreadsheet = pd.read_csv(spreadsheet_name)
new_spreadsheet.head()
```

```python
spreadsheet_name = "C:/Users/User/Downloads/Excel Spreadsheets/USC Center for Economic Development/Original Spreadsheet.csv"
old_spreadsheet = pd.read_csv(spreadsheet_name)
old_spreadsheet.head()
```

```python
new_spreadsheet["lat"] = 0
new_spreadsheet["long"] = 0
new_spreadsheet.head()
```

```python
hub_zones_name = "C:/Users/User/Downloads/Excel Spreadsheets/USC Center for Economic Development/Los_Angeles_Hubzones_TableToExcel.csv"
hub_zones = pd.read_csv(hub_zones_name)
hub_zones.head()
```

```python
valley_region_name = "C:/Users/User/Downloads/Excel Spreadsheets/USC Center for Economic Development/Valley Zip Codes.csv"
valley_region = pd.read_csv(valley_region_name)
valley_region.head()
```

# Identify all of the Valley Regions.

```python
for index1 in new_spreadsheet.index:
    for index2 in valley_region.index:
        if new_spreadsheet["BUSINESS ZIP CODE"][index1] == valley_region["ZIP CODE"][index2]:
            new_spreadsheet["VALLEY REGION"][index1] = "Yes"
```

# Recover all lost Data.

```python
# Spreadsheet is organized alphabetically by company name.
old_index = 0

# Keep looping until you get to the end of either spreadsheet.
while old_index < len(old_spreadsheet.index):
    
    # Checks to see if the company name exists in the old spreadsheet.
    if str(old_spreadsheet["Company_Name"][old_index]) in str(new_spreadsheet["COMPANY NAME"]):
        
        new_index = new_spreadsheet.index[new_spreadsheet["COMPANY NAME"]==old_spreadsheet["Company_Name"][old_index]]
        new_spreadsheet["lat"][new_index] = old_spreadsheet["Latitude"][old_index]
        new_spreadsheet["long"][new_index] = old_spreadsheet["Longitude"][old_index]
        new_spreadsheet["GEO MATCH"][new_index] = old_spreadsheet["Geo_Match_Level"][old_index]
        new_spreadsheet["CENSUS TRACT"][new_index] = "06037" + str(old_spreadsheet["Census_Tract"][old_index])
        new_spreadsheet["CA SENATE"][new_index] = old_spreadsheet["Senate District"][old_index]
        # Go to the next new index if match was found.
        new_index += 1
    
    old_index += 1
    # Check to see if we are loading data
    # print(old_index)
    
# We check the tail to see all of the data has been aggregated.
# We can determine this by the last two coluns, the lat and long.
new_spreadsheet.tail()
```

# Identify all of the Hub Zones.

```python
for index1 in new_spreadsheet.index:
    for index2 in hub_zones.index:
        if str(new_spreadsheet["CENSUS TRACT"][index1]) == "0" + str(hub_zones["Tract"][index2]):
            if hub_zones["Status_as_of_Jan_2018"][index2] == "Qualified":
                new_spreadsheet["HUB ZONE"][index1] = "Yes"
```
