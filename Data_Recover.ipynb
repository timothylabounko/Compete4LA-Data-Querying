```python
spreadsheet_name = "C:/Users/User/Downloads/Excel Spreadsheets/USC Center for Economic Development/Original Spreadsheet.csv"
old_spreadsheet = pd.read_csv(spreadsheet_name)
old_spreadsheet.head()
```

# Recover all lost Data.

```python
# Spreadsheet is organized alphabetically by company name.
old_index = 0

# Checks to see the number of copies.
copies = 0

new_ss_company_names = new_spreadsheet["COMPANY NAME"].tolist()

# Keep looping until you get to the end of either spreadsheet.
while old_index < len(old_spreadsheet.index):
    
    # Checks to see if the company name exists in the old spreadsheet.
    if str(old_spreadsheet["Company_Name"][old_index]) in new_ss_company_names:
                
        # We check company address to account for repeats in business names. This will be saved as the index we access from.
        new_index = new_ss_company_names.index(str(old_spreadsheet["Company_Name"][old_index]))
        prev_new_index = new_index
        
        while (new_ss_company_names[new_index] == new_ss_company_names[prev_new_index]):
            new_spreadsheet["TRANSACTION DATE YYYYMM"][new_index] = old_spreadsheet["Transaction_Date__YYYYMM_"][old_index]
            new_spreadsheet["CALL STATUS"][new_index] = old_spreadsheet["Call_Status_Description"][old_index]
            new_spreadsheet["IGID"][new_index] = old_spreadsheet["IGID"][old_index]
            # Increment both to account for the copies in company names.
            new_index += 1
            old_index += 1
            copies += 1
            
    
    if (copies < 1):
        old_index += 1
        # Check to see if we are loading data
        # print(old_index)
    # Reset copies to 0.          
    copies = 0
    
# We check the tail to see all of the data has been aggregated.
# We can determine this by the last two coluns, the lat and long.
new_spreadsheet.tail()
```
